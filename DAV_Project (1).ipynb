```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import PyPDF2
from collections import Counter
```

```python
pdf_path = "the kite runner.pdf"
def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as f:
        pdf_reader = PyPDF2.PdfReader(f)
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()
    return text
```

```python
def generate_word_cloud(text):
    wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```

```python
def generate_bar_plot(text):
    words = text.split()
    word_counts = Counter(words)
    most_common_words = word_counts.most_common(10)  # Get the top 10 most common words

    plt.figure(figsize=(10, 5))
    plt.bar(range(len(most_common_words)), [count for word, count in most_common_words], align='center')
    plt.xticks(range(len(most_common_words)), [word for word, count in most_common_words], rotation=45)
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.title('Top 10 Most Common Words')
    plt.show()
```

```python
def generate_pie_chart(text):
    words = text.split()
    word_counts = Counter(words)
    most_common_words = word_counts.most_common(10)  # Get the top 10 most common words

    labels = [word for word, count in most_common_words]
    counts = [count for word, count in most_common_words]
    e=[0,0.1,0.4,0.1,0,0.5,0.1,0.2,0,0.5]
    plt.figure(figsize=(8, 8))
    plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=140 ,explode=e,shadow=True,wedgeprops={"edgecolor" : "black",'linewidth': 1,'antialiased': True})
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.legend(labels)
    plt.title('Top 10 Most Common Words Distribution')
    plt.show()
```

```python
def main(pdf_path):
    text = extract_text_from_pdf(pdf_path)
    generate_pie_chart(text)


main(pdf_path)
```

```python
def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as f:
        pdf_reader = PyPDF2.PdfReader(f)
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()
    return text

def count_5_letter_words(text):
    words = text.split()
    five_letter_words = [word for word in words if len(word) == 5]
    return len(five_letter_words), five_letter_words[5:15]

def main(pdf_path):
    text = extract_text_from_pdf(pdf_path)
    num_5_letter_words, sample_words = count_5_letter_words(text)
    print(f"Number of 5-letter words: {num_5_letter_words}")
    print(f"Sample 5-letter words: {sample_words}")



main(pdf_path)
```

```python
from textblob import TextBlob
def extract_sentiment_words(pdf_path):
    text = extract_text_from_pdf(pdf_path)
    positive_words = []
    negative_words = []
    neutral_words = []

    for word in text.split():
        sentiment = TextBlob(word).sentiment.polarity
        if sentiment > 0:
            positive_words.append(word)
        elif sentiment < 0:
            negative_words.append(word)
        else:
            neutral_words.append(word)

    return positive_words, negative_words, neutral_words

def main(pdf_path):
    positive_words, negative_words, neutral_words = extract_sentiment_words(pdf_path)
    print("Positive words:", positive_words)
    print("Negative words:", negative_words)
    print("Neutral words:", neutral_words)


main(pdf_path)
```

```python
def count_sentiment_words(positive_words, negative_words, neutral_words):
    total_words = len(positive_words) + len(negative_words) + len(neutral_words)
    positive_percent = (len(positive_words) / total_words) * 100
    negative_percent = (len(negative_words) / total_words) * 100
    neutral_percent = (len(neutral_words) / total_words) * 100

    return positive_percent, negative_percent, neutral_percent

def main(pdf_path):
    positive_words, negative_words, neutral_words = extract_sentiment_words(pdf_path)
    positive_percent, negative_percent, neutral_percent = count_sentiment_words(positive_words, negative_words, neutral_words)

    # Plot pie chart
    labels = ['Positive', 'Negative', 'Neutral']
    sizes = [positive_percent, negative_percent, neutral_percent]
    colors = ['cyan', 'brown', 'orange']

    plt.figure(figsize=(5, 5))
    plt.pie(sizes,colors=colors,shadow = True, startangle = 90,  autopct='%1.1f%%',explode=[0,0.2,0.5],wedgeprops = {"edgecolor" : "black",'linewidth': 1,'antialiased': True})
    plt.legend(labels)
    plt.title('Sentiment Analysis Pie Chart')
    plt.show()



main(pdf_path)
```

```python
pdf_path = 'the kite runner.pdf'

# Extract text from PDF
pdf_text = extract_text_from_pdf(pdf_path)

# Perform sentiment analysis
blob = TextBlob(pdf_text)
sentimentpolarity = blob.sentiment.polarity
sentimentsubjectivity = blob.sentiment.subjectivity

# Display sentiment analysis results
print("Sentiment Polarity:", sentimentpolarity)
print("Sentiment Subjectivity:", sentimentsubjectivity)
```

```python
import matplotlib.pyplot as plt

# Define the sentiment polarity score
sentiment_polarity = sentimentpolarity

# Define the thresholds for sentiment categories
negative_threshold = -1
positive_threshold = 1

# Plotting
plt.figure(figsize=(8, 6))

# Plot sentiment categories
plt.axvline(x=negative_threshold, color='red', linestyle='--', label='Negative Sentiment')
plt.axvline(x=positive_threshold, color='green', linestyle='-.', label='Positive Sentiment')

# Plot sentiment polarity score
plt.plot(sentiment_polarity, 0, 'bo', markersize=10, label='Sentiment Polarity')

# Adding labels and legend
plt.xlabel('Sentiment Polarity Score')
plt.title('Sentiment Polarity Visualization')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
